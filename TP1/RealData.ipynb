```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
print(dir(iris))
```

```python
print(iris.DESCR)
```

```python
import pandas as pd

irisdf = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
print(irisdf)              # affiche contenu complet
print(irisdf.info())       # type, mémoire, count par colonne
print(len(irisdf))         # nombre de lignes
print(irisdf.size)         # nombre total de cellules
print(irisdf.index)        # index de lignes
print(irisdf.columns)      # noms de colonnes
print(irisdf.axes)         # index & colonnes
print(irisdf.shape)        # (150, 4)
print(irisdf.dtypes)       # type de données par colonne
print(irisdf.empty)        # False car dataframe non vide
```

```python
print(irisdf.head())
print(irisdf.head(8))
print(irisdf.tail())
print(irisdf['sepal length (cm)'])
print(irisdf[['sepal length (cm)','petal width (cm)']])

irisdf.iloc[0,0]
irisdf.iloc[0]
irisdf.iloc[5:15]
irisdf.iloc[:,1]
irisdf.iloc[:,1].head()
irisdf.iloc[:,[0,2]]
irisdf.iloc[5:20,[0,2]] 
```

```python
irisdf["target"] = iris.target
```

```python
print(irisdf.mean())
print(irisdf.describe())
print(irisdf.sort_values(by='sepal length (cm)'))
print(irisdf.sort_values(by='sepal length (cm)').head())
print(irisdf.isnull().sum())
print(irisdf.corr())
```

```python
irisdf["label"] = iris.target_names[irisdf["target"]]
```

```python
ax = irisdf.loc[irisdf['label']=="setosa"].plot.scatter(
    x='sepal length (cm)',
    y='sepal width (cm)',
    label='setosa',
    color='red'
)

irisdf.loc[irisdf['label']=="versicolor"].plot.scatter(
    x='sepal length (cm)',
    y='sepal width (cm)',
    label='versicolor',
    color='blue',
    ax=ax
)

irisdf.loc[irisdf['label']=="virginica"].plot.scatter(
    x='sepal length (cm)',
    y='sepal width (cm)',
    label='virginica',
    color='green',
    ax=ax
)
```
