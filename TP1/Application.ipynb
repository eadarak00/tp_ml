# 0.Importations

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
import pandas as pd
```

# 1. Génération des données

```python
X1, y1 = make_regression(n_samples=100, n_features=4, noise=10, random_state=42)
```

```python
print("1. Données générées :")
print(f"Shape de X1 : {X1.shape}") 
print(f"Shape de y1 : {y1.shape}")  
```

# 2. Analyse univariée des variables

```python
# Pour les 4 features
print("Min par colonne :", X1.min(axis=0))
print("Max par colonne :", X1.max(axis=0))
print("Moyenne par colonne :", X1.mean(axis=0))
print("Variance par colonne :", X1.var(axis=0))
print("Ecart-type par colonne :", X1.std(axis=0))

# pour la cible
print("\nTarget (y1) :")
print("Min :", y1.min())
print("Max :", y1.max())
print("Mean :", y1.mean())
print("Var :", y1.var())
print("Std :", y1.std())
```

# 3. Matrice de corrélation + nuages de points

```python
# Matrice corrélation
corr = np.corrcoef(X1, y1, rowvar=False)
print(corr)
```

```python
for i in range(4):
    plt.scatter(X1[:,i], y1)
    plt.xlabel(f"Feature {i}")
    plt.ylabel("Target")
    plt.title(f"Nuage des points : Feature {i} vs y1")
    plt.show()
```
