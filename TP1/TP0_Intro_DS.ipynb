# Introduction à la Science des Données - Python
Ce notebook présente étape par étape l'exploration de données avec sklearn, numpy, pandas et matplotlib.

## 1.Importation necessaires

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn import datasets
import pandas as pd
```

## 2.Génération de données synthétiques


```python
X, y = make_regression(n_samples=5, n_features=2, noise=10)
X, y
```

## 3.Exploration des données

```python
print(X)
print(X.dtype)
print(X.size)
print(len(X))
print(X.shape)
print(X[0])
print(X[0,0])
print(X[:,0])
```

## 4.Statistiques descriptives

```python
print(X.min(), X.max(), X.mean(), X.var(), X.std())
print(y.min(), y.max(), y.mean(), y.var(), y.std())
```

## 5.Analyses Bivariees

```python
np.corrcoef(X[:,0],y)
np.corrcoef(X,y,rowvar=False)
```

# 6.Graphique

```python
plt.scatter(X[:,0], y)
```

```python
plt.xlabel("Attribut2")
plt.ylabel("Cible")
plt.scatter(X[:,1], y,c='Red',label="Attribut2")
plt.legend()
plt.show()
```
